1.Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

PRIME NUMBER 

```python
num = int(input("Enter a Number: "))
is_prime = True
for i in range(2,num):
    if num%i == 0:
        is_prime = False
        break
if is_prime and num>1:
    print("Prime")
else:
    print("Not Prime")
```

2.Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message.

Product of Random Numbers

```python
import random


num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

answer = int(input("What is " + str(num1) + " × " + str(num2) + "? "))

if answer == num1 * num2:
    print("Correct answer!")
else:
    print("Wrong answer.")
    print("The correct answer is:", num1 * num2)
```

✅ Problem Statement (Question): Hospital Billing System

4.A hospital charges its patients based on the following:
•	Room charges per day: ₹2,000
•	Doctor consultation fee (flat): ₹1,500
•	Lab test charges: Based on the number of tests taken. Each test costs ₹300
•	Medicine charges: Total cost of medicines provided
•	Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied
Write a Python program to calculate the final bill for a patient who:
•	Stayed for 4 days
•	Had 3 lab tests
•	Had medicine charges of ₹2,400



```python
room_charge_per_day = 2000
doctor_fee = 1500
lab_test_cost_per_test = 300
days_stayed = 4

# Input from user
number_of_tests = int(input("Enter number of lab tests taken: "))
medicine_charges = float(input("Enter total medicine charges: ₹"))

# Calculations
room_charges = room_charge_per_day * days_stayed
lab_charges = number_of_tests * lab_test_cost_per_test

total_bill = room_charges + doctor_fee + lab_charges + medicine_charges


if total_bill > 10000:
    discount = total_bill * 0.10
else:
    discount = 0

final_bill = total_bill - discount

# Display bill details
print("\n----- Hospital Bill -----")
print("Room Charges: ₹", room_charges)
print("Doctor Fee: ₹", doctor_fee)
print("Lab Charges: ₹", lab_charges)
print("Medicine Charges: ₹", medicine_charges)
print("Total Bill (Before Discount): ₹", total_bill)
print("Discount: ₹", discount)
print("Final Bill to Pay: ₹", final_bill)
```
